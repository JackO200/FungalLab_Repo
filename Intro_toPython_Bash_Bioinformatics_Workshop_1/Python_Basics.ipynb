# <u>Ch. 2: Intro To Critical Packages & Conda Envs</u>
Now that we have your package managers installed, we can start talking about what packages are neccessary, as well as when to create a new conda environment vs what packages you should just have within your base env. We can think of environments like staying in hotels, they're not something we do everyday but they come with different perks, some are near the beach, some are near your family, others are just cool. Sometimes we stay in hotels for business trips, but regardless, environments serve a purpose, they contain packages that are specific to certain kinds of analyses or software. For example, there is a software called MetaSPADEs for bioinformatics, rather than have that in my base env, I can make a specific environment for that, and when I want to run MetaSPADES I can just activate that env and have access to MetaSPADEs and all of its capabilites. You can think of this really as a good way to stay organized. <br><br>
The following is some code that you can run in your terminal to get started with a conda env: <br><br>
**1. Check your env:** ``conda env list`` <br><br>
**2. Create a new env:** ``conda create -n YOUR_ENV_NAME python``, the addition of <span style="color: #FFAD69;">**python**</span> to the end of this says to conda I want you to install python when you create this environment, you could also do ``conda create -n YOUR_ENV_NAME python=3.12.9`` to specify the specific version of python or any package you want <br><br>
**3. Activate your conda env:** ``conda activate YOUR_ENV_NAME``<br><br>
**4. Install a new package:** ``conda install XXX_PACKAGE`` <br><br>
**5. Remove a package:** ``conda remove XXX_PACKAGE``<br><br>
**6. Remove a conda env:** ``conda remove -n YOUR_ENV_NAME --all`` <span style="color: #FFAD69;">**ONLY DO THIS IF YOU DO NOT WANT IT EVER AGAIN**</span> <br><br>

# <u> Ch. 4: Basic Data Viz with Seaborn</u>
Have you always wanted to be a pirate but get sea sick, well check out seaborn! Seaborn is a great python package for data visualization, and while to me it will never replace ggplot2 in R, it is great for making figures in python.

````python
# Distribution (Histogram) for the 'Phenolics' column
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Phenolics', kde=True, bins=30)
plt.title('Distribution of Phenolics')
plt.xlabel('Phenolics')
plt.ylabel('Frequency')
plt.show()
````

````python
# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Bench', y='AMFCount')
plt.title('AMFCount Distribution by Bench')
plt.xlabel('Bench')
plt.ylabel('AMFCount')
plt.show()
````

````python
# Scatter Plot with Relationship
sns.lmplot(data=df, x='Phenolics', y='Lignin', height=6, aspect=1.2, scatter_kws={'alpha':0.6, 'color': '#F5A6E6'}, line_kws={'color': '#BA2C73'})
plt.title('Relationship between NSC and Lignin')
plt.xlabel('NSC')
plt.ylabel('Lignin')
plt.show()
````

````python
# Pair Plot
# Selecting a subset of relevant columns
subset = df[['Phenolics', 'Harvest', 'Lignin']]

sns.pairplot(subset, kind='scatter', diag_kind='kde')
plt.suptitle('Pair Plot of Phenolics, Harvest, and Lignin', y=1.02)
plt.show()
````

````python
# Correlation Heatmap
# Compute the correlation matrix for numeric columns
# Select only numeric columns
df_num = df.select_dtypes(include=[np.number])
corr_matrix = df_num.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
````

# <u> Ch. 5: Loops & If Statements! </u>
<span style="color: #FFAD69;">**This is the most important part of this course**</span>, if you can write a good <span style="color: #FFAD69;">***efficient***</span> for loop, you will <span style="color: #FFAD69;"><u>***always***</u></span> have a job.
<br><br>
So what ***is*** a for loop?
<br><br>
A for loop is a fancy way of saying, I want to do a task or set of tasks, on a large sert of <span style="color: #FFAD69;">***sequential***</span> data. The important thing here is that the data needs to be something you want run one after the other, and you want the ***same*** thing done to every single file or object you are writing the loop on.
<br><br>
In terms of syntax a basic for loop has this structure:<br>
```
for i in (data):
    DO_THIS_THING
```
<br>
In this case, i = an individual file or object that you want to loop over. So you can do a single thing, or lots of things at the same time to a series of objects.
<br><br>
What is an if statement?
<br><br>
An if statement is a way to check IF something meets a certain condition. Here is the basic structure:
<br><br>

```
if condition:
    print("Condition is true")
```
<br><br>
We also can write an if/else statment, which has a structure of this:

```
if condition_A:
    print("Yay")
elif condition_B:
    print("Booo")
else:
    print("Whoops")
```

<br><br>
Where in an if else statment, or the above is an elif statement with three options. There is the ability to set numerous conditions to see what is doing what. You do not have to use print statmements, you can have something <span style="color: #FFAD69;">**happen**</span> to your data, but this is a simple example

````python
# Example of a very basic for loop
numbers = pd.Series(range(1, 100)) # Makes sure that the numbers are seen as a 1D array, rather than a list or something
## This loop will print every number in the list
for num in numbers:
    print(num)

# Can anyone tell me why it "skips" the last one?
````

````python
# Basic If/else Statements
for num in numbers:
    if num > 5:
        print(f"{num} is greater than 5")
    else:
        print(f"{num} number is lame!")
````

````python
# More complex loop
for index, row in df.iterrows():
    if row['Species'] == "Acsa":
        print("This is an Acsa")
    else:
        print("This is NOT an Acsa")
````
